```python
import pandas as pd 
import seaborn 
import matplotlib.pyplot as plt 
```

```python
#Kaggle URL https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows/

data = pd.read_csv(r'C:\Users\docto\OneDrive\Desktop\Python\IMBDMovieRankings\IMDBMovieData.csv')
```

```python
#For duplicate values in Data
if(data.duplicated().any() == True):
    data.drop_duplicates()
```

# Average IMDB Rating 

```python
imdbMean = format(data['Rating'].mean(), '.2f')
print("Average IMDB Movie Ranking: ", imdbMean)
```

# Compare IMDB Rating 


```python
#Year with Ratings 
data.groupby('Year')['Rating'].mean(), 
seaborn.barplot(x = 'Year', y = 'Rating', data = data)
plt.title("Rating Comparison by Year")
plt.show()

#Genre with Ratings 
data.groupby('Genre')['Rating'].mean()
seaborn.barplot(x = 'Genre', y = 'Rating', data = data)
plt.title("Rating Comparison by Genre")
plt.xticks(rotation = 45)
plt.show()
```

# IMDB Ratings Correlate with Revenue?

```python
seaborn.scatterplot(x = 'Rating', y = 'Revenue (Millions)', data = data)
plt.title("Revenue and Rating Correlation")
plt.show()
```
